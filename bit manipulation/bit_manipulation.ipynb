# **Binary Number Conversion - (1's & 2's compliment)**

$(7)_{10}$ - base 10 means the number is in decimal format.

$(00001)_2$ - base 2 means the number is in binary format

##### **Can we convert a number to it's binary equivalent? i.e $(7)_{10}$ ----> $(?)_{2}$**

```python
# Conversion - refer to bit manipulation notes in strivers section

# NOTE: divide the number by 2 until we reach either 1 or 0 and take out reminders
def convert_to_binary(num):
    binary = ""
    while num >= 1:
        div = num // 2
        binary = str(num - (2 * div)) + binary
        num = div

    return binary

print(f"binary equivalent of number 7 is {convert_to_binary(7)}")
# Time Complexity - Log N (actually log 2 N to be specific)
# Space Complexity - Log N (actually log 2 N to be specific)
```

##### **Can we convert a binary to it's decimal equivalent? i.e $(111)_{2}$ ----> $(?)_{10}$**

```python
# Conversion - refer to bit manipulation notes in strivers section

# NOTE: Indexing starts from right to left. Multiple each bit with 2 raised to the power of index

def convert_to_decimal(binary):
    decimal = 0
    for idx in range(len(binary)):
        decimal += (int(binary[-idx + 1]) * ( 2**idx))

    return decimal

print(f"decimal equivalent of binary 111 is {convert_to_decimal("111")}")
# Time Complexity - O(n)
# Space Complexity - O(n)
```
