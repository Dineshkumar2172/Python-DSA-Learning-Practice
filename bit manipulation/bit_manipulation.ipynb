# **Binary Number Conversion - (1's & 2's compliment)**

$(7)_{10}$ - base 10 means the number is in decimal format.

$(00001)_2$ - base 2 means the number is in binary format

##### **Can we convert a number to it's binary equivalent? i.e $(7)_{10}$ ----> $()_{2}$**

```python
import math

# Conversion - refer to bit manipulation notes in strivers section

# divide the number by 2 until we reach either 1 or 0 and take out reminders
def convert_to_binary(num):
    binary = ""
    while num >= 1:
        div = num // 2
        binary = str(num - (2 * div)) + binary
        num = div

    return binary

convert_to_binary(7)
```
