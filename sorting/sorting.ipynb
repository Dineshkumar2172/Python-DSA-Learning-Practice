## **Selection Sort**

step 1: pick minimum element from the unsorted section of an array

step 2: swap the picked element at the end of sorted section with the minimum element picked.

step 3: now expand the explored are and continue search and expansion.

```python
arr = [13, 46, 24, 52, 20, 9]

def selection_sort(arr):
    def swap(idx1, idx2):
        arr[idx1], arr[idx2] = arr[idx2], arr[idx1]

    for idx in range(0, len(arr)):
        minimum_index = idx
        for inner_idx in range(idx, len(arr)):
            if arr[inner_idx] < arr[minimum_index]:
                minimum_index = inner_idx

        swap(idx, minimum_index)

selection_sort(arr)
print(arr)
```

## **Bubble Sort** - swap all adjacent elements n times

```python
arr = [13, 46, 24, 52, 20, 9]

def bubble_sort(arr):
    for idx in range(len(arr)):
        for idx2 in range(len(arr) - 1):
            if arr[idx2] > arr[idx2+1]:
                temp = arr[idx2]
                arr[idx2] = arr[idx2 + 1]
                arr[idx2 + 1] = temp

bubble_sort(arr)
print(arr)
```

## **Insertion Sort**

1. Pick ad element and compare it with it's previous element.
2. If it's lesser than previous element, then swap it.
3. Now check again, if again it's lesser than new previous element, then swap it.
4. Continue till we find it's correct position.

```python
arr = [13, 46, 24, 52, 20, 9]

def insertion_sort(arr):
    def swap(idx1, idx2):
        temp = arr[idx1]
        arr[idx1] = arr[idx2]
        arr[idx2] = temp
    
    for idx in range(1, len(arr)):
        temp_idx = idx
        while arr[temp_idx] < arr[temp_idx-1] and temp_idx > 0:
            swap(temp_idx, temp_idx-1)
            temp_idx -= 1

insertion_sort(arr)
print(arr)
```
